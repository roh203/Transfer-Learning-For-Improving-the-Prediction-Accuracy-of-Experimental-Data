```python
import pandas as pd
data = pd.read_csv('/content/Overall Catalytic Activity_CuMgO_Hydrothermal.csv')
```

```python
data
```

```python
data = data.drop(['Carbon Species','STY (gmethanol/gcat.hr)','STY (mol/gcat.hr)','STY (mmol/gcat.hr)','x','STY(g Catkg-1 h-1 )'], axis = 1)
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_sc = scaler.fit_transform(data.to_numpy())
data_sc = pd.DataFrame(data_sc, columns = ['BET','ASA','Temp(C)','Pre','WHSV','CO2 C','Sel',])
```

```python
x = data_sc.iloc[:,:-2]
y = data_sc.iloc[:,-2]
```

```python
x
```

```python
y_pred = new_model.predict(x)
y_pred
```

```python
y
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
for layer in new_model.layers:
  layer.trainable = False
new_model.summary()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
from tensorflow.keras import layers
```

```python
import tensorflow as tf
```

```python
new_model.summary()
```

```python
ultra_model = Sequential()
ultra_model.add(Dense(5, input_dim=5, kernel_initializer='normal', activation='relu'))
ultra_model.add(Dense(5, activation = 'relu'))
ultra_model.add(new_model)
ultra_model.add(Dense(216,activation='relu'))
ultra_model.add(Dense(1, activation='linear'))
```

```python
ultra_model.summary()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=.2, random_state = 113)
```

```python
losses = tf.keras.losses.MeanAbsoluteError(reduction="auto", name="mean_squared_error")
ultra_model.compile(loss=losses, optimizer='adam')
history=ultra_model.fit(x_train, y_train, epochs=1000, batch_size=20, verbose=1, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
